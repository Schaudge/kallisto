Print BUS file contents and generate count matrix for all test files:

```python
for frame in ["frame0", "frame1", "frame2", "frame3", "frame4", "frame5"]:
    print(frame)
    
    !bustools sort -o $out_folder/$frame/output_sorted.bus $out_folder/$frame/output.bus
    
    !bustools text -pf $out_folder/$frame/output.bus
    
    !bustools count \
        -o $out_folder/$frame/bustools_count/ \
        -g $index/cfc_palmdb_t2g.txt \
        -e $out_folder/$frame/matrix.ec \
        -t $out_folder/$frame/transcripts.txt \
        $out_folder/$frame/output_sorted.bus
```
