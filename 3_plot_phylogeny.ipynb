```python
def flatten(l):
    return [item for sublist in l for item in sublist]
```

```python
## Clean up transctipt numbers
# Remove \t and \n
contents = [i.split('\t')[1].split('\n')[0] for i in contents]

# Not doing the following removes all multimappings
# # Split list of numbers into separate strings
# contents = [i.split(',') for i in contents]
# # TO DO: ONly flatten if it is a list of lists!!!
# contents = flatten(contents)

contents
```

```python
# Count number of cells expressing the transcript
counts = {}
for i in contents:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1
counts
```

```python
# Get corresponding phylogenies for each number from look-up file
path_to_tsv = "/Users/lauraluebbert/OneDrive - California Institute of Technology/Pachter_lab/comma-free/palmdb/2021-03-14/u_tax.tsv"
phylogeny_data = pd.read_csv(path_to_tsv,sep='\t')

# Remove "u"s from transcript IDs 
phylogeny_data["clean_label"] = phylogeny_data["Label"].str.split("u").str[1]

# Only keep transcripts found in data
phylogeny_data = phylogeny_data[phylogeny_data["clean_label"].isin(contents)]

# Add number of cells expressing transcript
phylogeny_data['num_cells'] = phylogeny_data['clean_label'].map(counts)
phylogeny_data
```

```python
# Remove non-relevent columns
phylogeny_data_clean = phylogeny_data.drop(['Label', 'clean_label', 'num_cells'], axis=1)
# Add "root" node
phylogeny_data_clean.insert(0, "root", "root")

phylogeny_data_clean
```

```python
node_to_children = {}
# Setting arbritrary branch length 1
branch_length = 1

# Iterate over dataframe row-wise, assuming that every row stands for one complete branch of the tree
for row in phylogeny_data_clean.itertuples():
    # Remove index at position 0 and elements that contain no child (".")
    row_list = [element for element in row[1:] if element != "."]

    for i in range(len(row_list)-1):
        if row_list[i] in node_to_children.keys():
            # Check if parent entry already existing 
            if row_list[i+1] in node_to_children[row_list[i]].keys():
                # If already exists --> next
                continue
            else:
                # If entry does not exist --> update dict and add the connection
                node_to_children[row_list[i]].update({row_list[i+1]:branch_length})
        else:
            # Add the branching point
            node_to_children[row_list[i]] = {row_list[i+1]:branch_length}

node_to_children
```

```python
def newickify(node_to_children, root_node) -> str:
    visited_nodes = set()

    def newick_render_node(name, distance: float) -> str:
        assert name not in visited_nodes, "Error: The tree may not be circular!"

        if name not in node_to_children:
            # Leafs
            return F'{name}:{distance}'
        else:
            # Nodes
            visited_nodes.add(name)
            children = node_to_children[name]
            children_strings = [newick_render_node(child, children[child]) for child in children.keys()]
            children_strings = ",".join(children_strings)
            return F'({children_strings}){name}:{distance}'

    newick_string = newick_render_node(root_node, branch_length) + ';'

    # Ensure no entries in the dictionary are left unused.
    assert visited_nodes == set(node_to_children.keys()), "Error: some nodes aren't in the tree"

    return newick_string
```

```python
nw_string = newickify(node_to_children, root_node='root')
nw_string
```

```python
import newick
tree = newick.loads(nw_string)[0]
print(tree.ascii_art())
```

```python
from ete3 import Tree, faces, AttrFace, TextFace, TreeStyle, NodeStyle, CircleFace

def my_layout(node):
    fsize = 9
    if node.is_leaf():
        # If terminal node, draws its name
        name_face = AttrFace("name")
    else:
        # If internal node, draws label with smaller font size
        name_face = AttrFace("name", fsize=fsize)
    # Adds the name face to the image at the preferred position
    faces.add_face_to_node(name_face, node, column=0, position="branch-right")

    # Add spheres that represent the number of cells expressing each virus
    if "weight" in node.features:
        # # Creates a sphere face whose size is proportional to node'sfeature "weight"
        # C = CircleFace(radius=node.weight, color="RoyalBlue", style="sphere")
        # C.opacity = 0.3
        # C.hz_align = 2
        
        # # And place as a float face over the tree
        # faces.add_face_to_node(C, node, 0, position="float-behind")

        # Add number of cells to node
        text_face = TextFace(f"# Cells expressing: {node.weight}", fsize=fsize)
        faces.add_face_to_node(text_face, node, column=0, position="branch-right")

        # Add node the size of the node weight
        nstyle = NodeStyle()
        nstyle["fgcolor"] = "LightBlue"
        nstyle["size"] = node.weight
        node.set_style(nstyle)

ts = TreeStyle()
# Do not add leaf names automatically
ts.show_leaf_name = False
# Use my custom layout
ts.layout_fn = my_layout

t = Tree(nw_string, format=1)

i = 0
for n in t.traverse():
     # Remove standard node spheres
     nstyle = NodeStyle()
     nstyle["fgcolor"] = None
     nstyle["size"] = 0
     n.set_style(nstyle)

     if n.is_leaf():
          # Add weights that represent the number of cells expressing each virus
          # http://etetoolkit.org/docs/latest/tutorial/tutorial_drawing.html
          n.add_features(weight=phylogeny_data['num_cells'].values[i])
          i+=1

# Tell ETE to use your custom Tree Style
t.show(tree_style=ts)
```
